```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y=make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
```

```python
#Applying OLS
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
#applying gradient descent stepwise
#lets assume the starting value for he intercept b=0
y_pred=((78.35063668*X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red', label='OLS')
plt.plot(X,y_pred,color='green', label='b=0')
plt.legend()
plt.show()
```

```python
m=78.35
b=0

loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
lr=0.1
step_size=lr*loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred1=((78.35063668*X)+b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red', label='OLS')
plt.plot(X,y_pred,color='green', label='b=0')
plt.plot(X,y_pred1,color='Yellow', label='b=20.927')
plt.legend()
plt.show()
```

```python
b=100
m=78.35
lr=0.1
epochs=10
plt.figure(figsize=(15,15))
for i in range(epochs):
  loss_slope=-2*np.sum(y-m*X.ravel()-b)
  b=b-(lr*loss_slope)
  y_pred=m*X+b

  plt.plot(X,y_pred)
plt.scatter(X,y)
```
